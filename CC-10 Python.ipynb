```
#Task - 1 (Lambda Functions)
given_list = [['C', 'L'], ['A', 'R', 'U'], ['S', 'W', 'A', 'Y']]
def biggest_list(x):      
   longestList = max(x, key = lambda a: len(a))
   maxLength = max(map(len, x))
   return longestList, maxLength
print(biggest_list(given_list))
```

```
#Task - 2 (Defining Functions)
def divisible(x,y):
    try:
        return "divisible" if x%y == 0 else "not divisible"
    except ZeroDivisionError:
        print("error")

num = float(input("Enter the number you want to learn if divisible"))
divider = float(input("Enter the divider number in order to see if divisible or not divisible"))
divisible(num, divider)
```
