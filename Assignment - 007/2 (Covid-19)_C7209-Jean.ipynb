```
age = input("Are you a cigarette addict older than 75 years old? Please answer the questions just as 'Yes' or 'No'").title().strip()
chronic = input("Do you have a severe chronic disease? Please answer the questions just as 'Yes' or 'No'").title().strip()
immune = input("Is your immune system too weak? Please answer the questions just as 'Yes' or 'No'").title().strip()
if age == "Yes":
    age = True
else:
    age = False
if chronic == "Yes":
    chronic = True
else:
    chronic = False
if immune == "Yes":
    immune = True
else:
    immune = False
if age or chronic or immune:
    print("You are in risky group")
else:
    print("You are not in risky group")
```
