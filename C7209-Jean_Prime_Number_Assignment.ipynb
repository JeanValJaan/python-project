```
number = input("Enter a positive integer in order to check if it is an prime number or not.")
if int(number) == 2: 
    print(f"{number} is a prime number")
else:
    for i in range(2, int(number)):
        if int(number) % i == 0 :
          print(f"{number} is not a prime number")
          break
        else:
          print(f"{number} is a prime number")
          break


    
```
