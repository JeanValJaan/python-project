```
limit = int(input("Enter the upper limit that you want to list the prime numbers"))
prime_list = []
for num in range(2, limit + 1):    
    for i in range(2,num):  
        if (num % i) == 0:  
            break  
    else:  
        prime_list.append(num)

print(prime_list)
```
