```
def fibonacci(n):
   
    if n < 0:
        print("Incorrect input")
 
    elif n == 0:
        return 0
 
    elif n == 1 or n == 2:
        return 1
 
    else:
        return fibonacci(n-1) + fibonacci(n-2)
 
i = 1
fibo_list = []
limit = 56
while fibonacci(i) < limit :
    fibo_list.append(fibonacci(i))
    i += 1
print(f"fibonacci numbers less than {limit} are : {fibo_list}")
```
